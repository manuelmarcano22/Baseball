# Search for what happend at the 2-2

````python
duckMookie = mookiegamesordered[(~mookiegamesordered['events'].isna()) 
                                & (mookiegamesordered.balls == 2) 
                                & (mookiegamesordered.strikes == 2)
                               & (mookiegamesordered.outs_when_up == 2)]
````
