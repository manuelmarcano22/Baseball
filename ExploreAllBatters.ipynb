```python
allbat2019 = pd.read_csv('AllBatters2019.csv')
```

```python
duckMookies = allbat2019[(allbat2019.player_name == 'Mookie Betts') 
                  & (~allbat2019['events'].isna()) 
                  & (allbat2019.balls == 2) 
                  & (allbat2019.strikes == 2)
                  & (allbat2019.outs_when_up == 2)]
```

```python
ducks = allbat2019[
                  (~allbat2019['events'].isna()) 
                  & (allbat2019.balls == 2) 
                  & (allbat2019.strikes == 2)
                  & (allbat2019.outs_when_up == 2)]
```

```python
eventsnonan = allbat2019[(~allbat2019['events'].isna()) ]
eventsingle = allbat2019[allbat2019['events'] == 'single' ]
```

```python
tablecounts = eventsnonan.groupby(['balls','strikes','outs_when_up']).size().reset_index().rename(columns={0:'count'})
```

```python
tablecounts['percentage'] = tablecounts['count']/tablecounts['count'].sum()*100.
tablecounts.sort_values(by=['count'])
```

```python
tablecountsduck = ducks.groupby(['balls','strikes','outs_when_up','events']).size().reset_index().rename(columns={0:'count'})
tablecountsduck = tablecountsduck.sort_values(by=['count'])
```

```python
tablecountsduck['percentage'] = (tablecountsduck['count']/tablecountsduck['count'].sum())*100.
tablecountsduck['cum_sum_percent'] = (tablecountsduck['count'].cumsum()/tablecountsduck['count'].sum())*100.
tablecountsduck
```
